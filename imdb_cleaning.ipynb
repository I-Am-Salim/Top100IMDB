- Some dates are in words rather than numbers

**Column: Country**
- Some typos in country spellings

**Column: Income**
- Are of type str instead of Int

**Column: Votes**

**Column: Genre**
- Multiple value column

```python
df.columns
print(df['Unnamed: 8'].isna().all())
df.drop(columns='Unnamed: 8', inplace=True)
```

```python
mask = df.isna().all(axis=1)
df[mask]
```

```python
df.dropna(how ='all', inplace=True)
```

```python
df.columns = ['title_id', 'original_title', 'release_year', 'genre', 'duration', 'country', 'content_rating', 'director','income','votes','score']
df
```

```python
df.duplicated().sum()
```

**Initial Cleaning**
- Unnamed/Empty Column and Row dropped
- Renamed the column names to clean appropriate snake-case versions
- No duplicates found so none dropped

```python
df['country'].value_counts()
```

```python
country_fix = {'New Zesland': 'New Zealand', 'New Zeland':'New Zealand', 'US.':'USA', 'US':'USA', 'Italy1':'Italy', 'West Germany':'Germany'}
df['country'] = df['country'].replace(country_fix)
```

```python
df['country'].value_counts()
```

```python
df['duration'] = df['duration'].str.replace(r"[^0-9]", "", regex=True)
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')
df['duration'] = df['duration'].fillna(df['duration'].median())
```

```python
df['duration'].describe()
```

**Country and Duration**
- Country misspellings fixed but ensuring each country has one correct spelling
- Null/Garbage values removed from duration and replaced by the median 

```python
df['content_rating'].unique()
```

```python
ratings = {'Not Rated':'Unrated'}
df['content_rating'] = df['content_rating'].replace(ratings)
```

```python
df['content_rating'].unique()
```

```python
df['content_rating'] = df['content_rating'].fillna('Unknown')
```

```python
df['content_rating'].value_counts(dropna=False)
```

**Content Rating**
- Not Rated and Unrated combined into a single category Unrated
- Approved kept as a historical category
- All empty ratings 

```python
df['income'] = df['income'].str.replace('o','0')
df['income'] = df['income'].str.replace(r"[^0-9]","", regex=True)
df['income'] = pd.to_numeric(df['income'], errors = 'coerce')
df.loc[df['original_title'] == '12 Angry Men', 'income'] = 2877755
```

```python
df.head()
df.info()
```

**Income**
- Dollar signs removed
- Typos of 0 as o fixed
- Column converted to integer

```python
df['votes'] = df['votes'].str.replace('.','')
df['votes'] = pd.to_numeric(df['votes'], errors='coerce')
df['votes']
```

```python
df['score'] = df['score'].replace(r"[,:]", '.', regex=True)
df['score'] = df['score'].replace(r"[^0-9^\.]", "", regex=True)
df['score'] = df['score'].replace(r"\.\.", '.', regex=True)
df['score']= pd.to_numeric(df['score'], errors='coerce')
df['score'] = df['score'].fillna(df['score'].median())
df['score'].value_counts()
```

**Score**
- All non decimal punctuation and repeated decimals replaced with single decimals
- All non numeric characters removed
- Column converted to floats and null value replaced with the median of the column


```python
df['release_year'].unique()
parsed = pd.to_datetime(df['release_year'], errors = 'coerce', format='mixed')
df[parsed.isna()]['release_year']
```

```python
df[parsed.isna()]['original_title']
```

```python
df.loc[df['original_title']== 'Sunset Blvd.', 'release_year'] = '06-03-1951'
df.loc[df['original_title']== 'Scarface', 'release_year'] = '01-01-1984'
df.loc[df['original_title']== 'Taxi Driver', 'release_year'] = '01-01-1976'
df['release_year'] = pd.to_datetime(df['release_year'], errors = 'coerce', format='mixed')
check = df['release_year'].dt.year.between(1920,2024)
df[~check]
```

```python
df['year'] = df['release_year'].dt.year
```

**Release Year**
- All string dates are now in a datetime format
- Invalid dates were fixed manually 
- Out of bound date found which was fixed by looking up the correct date on IMDB
